```python
plt.figure(figsize=(10,6))
label_count=df["Survived"].value_counts().values
label=df["Survived"].value_counts().index

plt.subplot(1,2,1)
explode=(0,0.1)
colors=["orange","green"]
labels=["No","Yes"]
plt.pie(x=label_count,labels=labels,autopct="%1.1f%%2",shadow=True,radius=1,colors=colors,explode=explode)
plt.title("survival distribution")


plt.subplot(1,2,2)
colors=["yellow","red"]
sns.countplot(x=df["Survived"],palette=colors)
plt.xticks([0,1],["No","Yes"])
plt.title("survival count")

plt.subplots_adjust(wspace=0.5)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```
